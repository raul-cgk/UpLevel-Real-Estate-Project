### Step 4: Read both CSVs as DataFrames

### Step 6: Replace missing data in DataFrames
What did you notice? It seems like most of the columns in both DataFrames are not empty. This is good because you won't have much headache trying to fill the NA data.



This is what you should see if you did things correctly. It's not a lot of data that's missing, but it's always good to see what we can do instead of throwing it away. 

What we can do with the missing data in H1's column is
1. we can replace the NA values with "Unknown" 

For H2, we have two columns with missing data. What we can do is:
1. replace the NA values with "Unknown" for the same column you found in H1
2. replace the NA values with 0 in the column containing only 4 missing data

<font color = 'red'>Don't forget to rerun Step 5 to confirm that you've dealt with the missing data.</font>

<details>
    <summary><font color = 'green'>Click once for hint</font></summary>
    <div>
        <strong>Google "replace na in column pandas"</strong>
    </div>
</details>

```python
# Step 6a: Replace the NA values in H1 DataFrame with "Unknown"

H1["Country"].fillna("Unknown", inplace = True)
```

```python
# Step 6b: Replace the NA values in the same H1 column in the H2 DataFrame with "Unknown"

H2["Country"].fillna("Unknown", inplace = True)
```

```python
# Step 6c: Replace the NA values in the column containing 4 missing data with 0

H2["Children"].fillna(0, inplace = True)
```

```python
# Step 6d: Check to see if you've filled all the NA values

H1.isnull().sum().sum()

H2.isnull().sum().sum()
```

### Step 10: Clean 'Meal', 'ReservedRoomType', 'AssignedRoomType', and 'DepositType' for H1
There's a lot of whitespace in the four columns, so let's remove that. 

For example, you are turning '       BB           ' in the 'Meal' column into 'BB'. 

<details>
    <summary><font color = 'green'>Click once for cleaning hint</font></summary>
    <div>
        <strong>Google "pandas remove spaces from column values" </strong>
    </div>
</details>

```python
# Step 10a: Remove spaces in values in Meal for H1

H1["Meal"] = H1["Meal"].str.strip()
```

```python
# Step 10b: Remove spaces in values in ReservedRoomType for H1

H1["ReservedRoomType"] = H1["ReservedRoomType"].str.strip()
```

```python
# Step 10c: Remove spaces in values in AssignedRoomType for H1

H1["AssignedRoomType"] = H1["AssignedRoomType"].str.strip()
```

```python
# Step 10d: Remove spaces in values in DepositType for H1

H1["DepositType"] = H1["DepositType"].str.strip()
```

### Step 11: Repeat Step 10 for H2
Clean the same for H2 as well.

```python
# Step 11a: Remove spaces in values in Meal for H2
H2["Meal"] = H2["Meal"].str.strip()
# Step 11b: Remove spaces in values in ReservedRoomType for H2
H2["ReservedRoomType"] = H2["ReservedRoomType"].str.strip()
# Step 11c: Remove spaces in values in AssignedRoomType for H2
H2["AssignedRoomType"] = H2["AssignedRoomType"].str.strip()
# Step 11d: Remove spaces in values in DepositType for H2
H2["DepositType"] = H2["DepositType"].str.strip()
```

### Step 12: Check the first row again, i.e. repeat Step 9
Let's see if Steps 10 and 11 helped in cleaning.

If you did it right, the values should all look normal, i.e. do not contain excess spaces.
